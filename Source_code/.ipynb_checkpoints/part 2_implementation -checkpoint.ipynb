```python
'''
Rate of net jobs created from expanding/contracting and opening/closing establishments during the last 12 months (NET_JOB_CREATION_RATE)
Year (time)
Number of establishments (ESTAB)
Number of employees (EMP)
'''
#rename these columns 
# key = old name
# value = new name

dict = {'Rate of net jobs created from expanding/contracting and opening/closing establishments during the last 12 months (NET_JOB_CREATION_RATE)': 'Net Job Creation',
        'Year (time)': 'year',
        'Number of establishments (ESTAB)': 'Number of Establishments',
        'Number of employees (EMP)': 'Number of employees',
        'Rate of jobs lost from closing establishments during the last 12 months (JOB_DESTRUCTION_RATE_DEATHS)':
        'Rate of Job loss'}
 
# call rename () method
df_1.rename(columns=dict,
          inplace=True)
```
